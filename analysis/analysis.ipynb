```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("videos.csv")
df.head()
```

```python
df['engagement_rate'] = (df['likes']+df['comments']+df['shares']+df['collections'])/df['views']
df.head()
```

```python
df["publish_date"] = pd.to_datetime(df["publish_date"])
```

```python
collection_date = pd.to_datetime("2026-06-03")

df["age"] = (
    collection_date - df["publish_date"]
).dt.days

df["tp"] = df["views"] / np.log1p(df["age"]+1)

df.head()
```

```python
df['tp_rank'] = df['tp'].rank(ascending = False)
df['er_rank'] = df['engagement_rate'].rank(ascending = False)
df['avg_rank_score'] = (df['tp_rank']+df['er_rank'])/2 
df['avg_rank'] = df['avg_rank_score'].rank()

df.head()
```
