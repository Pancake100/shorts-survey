```python
import json
import glob
import pandas as pd
```

```python
json_files = glob.glob("experiment_data/*.json")

print("Number of JSON files:", len(json_files))

participant_rows = []
evaluation_rows = []
```

```python
for file in json_files:

    with open(file, "r", encoding="utf-8") as f:
        data = json.load(f)

    payload = data["payload"]

    participant_id = data["participant_id"]
    group = data["group"]

    # Participant Data
    pre = payload["preliminary_questionnaire"]

    participant_row = {
        "participant_id": participant_id,
        "group": group,
        "age": pre.get("age"),
        "gender": pre.get("gender"),
        "country_region": pre.get("country_region"),
        "native_language": pre.get("native_language"),
        "education_level": pre.get("education_level"),
        "short_video_frequency": pre.get("short_video_frequency"),
        "platforms": ", ".join(pre.get("platforms", [])),
        "usual_session_length": pre.get("usual_session_length"),
        "usual_video_types": ", ".join(
            pre.get("usual_video_types", [])
        ),
        "used_short_videos_for_learning":
            pre.get("used_short_videos_for_learning"),
        "learning_short_video_frequency":
            pre.get("learning_short_video_frequency"),
        "educational_video_usefulness":
            pre.get("educational_video_usefulness"),
        "topic_preference_1":
            pre.get("topic_preference_ranking", [None, None, None])[0],
        "topic_preference_2":
            pre.get("topic_preference_ranking", [None, None, None])[1],
        "topic_preference_3":
            pre.get("topic_preference_ranking", [None, None, None])[2],
        "english_comfort_level":
            pre.get("english_comfort_level")
    }

    participant_rows.append(participant_row)

    # Video Evaluation Data
    for response in payload["responses"]:

        evaluation = response["evaluation"]

        visual = evaluation["visual_impression"]
        content = evaluation["content_impression"]
        audio = evaluation["audio_impression"]
        overall = evaluation["overall_evaluation"]

        evaluation_row = {
            "participant_id": participant_id,
            "group": group,

            "video_id": response["video_id"],
            "video_title": response["video_title"],
            "topic": response["video_topic"],
            "video_order": response["video_order_index"],

            **visual,

            **content,

            **audio,

            **overall
        }

        evaluation_rows.append(evaluation_row)
```

```python
participant_df = pd.DataFrame(participant_rows)
evaluation_df = pd.DataFrame(evaluation_rows)

print("Participants:", participant_df.shape)
print("Evaluations:", evaluation_df.shape)
```
